```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt
```

```python
model = YOLO(r'D:\GitHub\safezone-ai\models\best.pt')
```

```python
image = cv2.imread(r"C:\Users\mohan\Downloads\inf.jpg")
results = model(image, show=False)[0]
for result in results.boxes.data.tolist():
    classNames = results.names
    x1, y1, x2, y2, score, class_id = result
    
    cv2.rectangle(image, (int(x1),int(y1)), (int(x2),int(y2)), (0,0,255), 4)
    
    cv2.rectangle(image, (10,10), (300,80), (255,0,0), -1)
    cv2.putText(image, classNames[class_id], (40,50), cv2.FONT_HERSHEY_DUPLEX, 1, (0,255,0), 1)
    print(classNames[class_id])
    plt.imshow(image[:,:,::-1])
```
